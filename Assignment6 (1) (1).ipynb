```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# Variables
x = np.arange(0, 100, 1)
y = 2 * x
z = x ** 2
```

```python
# Plotting (X,Y) on an axes ax
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('Days')
ax.set_ylabel('Growth')
ax.set_title('Experiment')
plt.show()
```

```python
# Plotting (X,Y) on two axes ax1 and ax2
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, y)
ax2.plot(x, y)
plt.show()
```

```python
# Create two subplots: 1 row 2 columns on a canvas that is 10 inches wide and 3 inches in height
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))
ax1.plot(x, y, 'r-', linewidth=3)
ax2.plot(x, z, 'b--', linewidth=1)
plt.show()
```

```python
# Import the seaborn library and load the built-in seaborn data set called “titanic”
titanic = sns.load_dataset('titanic')
print(titanic.head())
```

```python
# Create a joint plot that will chart the fare and age data columns
sns.jointplot(x='fare', y='age', data=titanic)
plt.show()
```

```python
# Create a distribution plot that will chart the fare column with 30 bins using color red
sns.histplot(titanic['fare'], bins=30, color='red')
plt.show()
```

```python
# Create a box plot that has the class column along the x-axis and the age column along the y-axis
sns.boxplot(x='class', y='age', data=titanic)
plt.show()
```

```python
# Create a count plot that charts the sex column
sns.countplot(x='sex', data=titanic)
plt.show()
```
